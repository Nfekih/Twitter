```python
import pickle
import twitter 
import json
import os
```

```python
if not os.path.exists('secret_twitter_credentials.pkl'):
    Twitter={}
    Twitter['Consumer Key'] = '########################################################'
    Twitter['Consumer Secret'] = '#####################################################'
    Twitter['Access Token'] = '########################################################'
    Twitter['Access Token Secret'] = '#################################################'
    with open('secret_twitter_credentials.pkl','wb') as f:
        pickle.dump(Twitter, f)
else:
    Twitter=pickle.load(open('secret_twitter_credentials.pkl','rb'))
```

```python
auth = twitter.oauth.OAuth(Twitter['Access Token'],
                           Twitter['Access Token Secret'],
                           Twitter['Consumer Key'],
                           Twitter['Consumer Secret'])
twitter_api = twitter.Twitter(auth=auth)
print(twitter_api)
```

```python
# woeid.rosselliot.co.nz/lookup/

WORLD_WOE_ID = 1
NEWYORK_WOE_ID = 2459115
PARIS_WOE_ID = 615702

world_trends = twitter_api.trends.place(_id=WORLD_WOE_ID)
ny_trends = twitter_api.trends.place(_id=NEWYORK_WOE_ID)
paris_trends = twitter_api.trends.place(_id=PARIS_WOE_ID)
```

```python
print((json.dumps(world_trends[:2], indent=1)))
```

```python
print((json.dumps(ny_trends[:2], indent=1)))
```

```python
print((json.dumps(paris_trends[:2], indent=1)))
```

```python
trends_set = {}
trends_set['world'] = set([trend['name'] 
                        for trend in world_trends[0]['trends']])

trends_set['new_york'] = set([trend['name'] 
                     for trend in ny_trends[0]['trends']]) 

trends_set['paris'] = set([trend['name'] 
                     for trend in paris_trends[0]['trends']]) 
```

```python
for loc in ['world','new_york','paris']:
    print(('~'*20,loc))
    print((','.join(trends_set[loc])))
```

```python
print(( '~'*20,'intersection of world and new_york'))
print((trends_set['world'].intersection(trends_set['new_york'])))
```

```python
print(( '~'*20,'intersection of new_york and paris'))
print((trends_set['new_york'].intersection(trends_set['paris'])))
```

### Search with a specific #hashtag

```python
q = '#Nike' 
number = 100

# See https://dev.twitter.com/docs/api/1.1/get/search/tweets

search_results = twitter_api.search.tweets(q=q, count=number)
statuses = search_results['statuses']
```

```python
print(len(statuses))
print(statuses)
```

Twitter often returns duplicate results.


To check if there are duplicate results, print out len of the list pre- and post- filtering.

```python
len(statuses)
```

```python
all_text = []
filtered_statuses = []
for s in statuses:
    if not s["text"] in all_text:
        filtered_statuses.append(s)
        all_text.append(s["text"])
statuses = filtered_statuses   
```
